```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Coordenadas del sitio de lanzamiento
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# Coordenadas de una ciudad cercana (por ejemplo, Cocoa, FL)
city_lat = 28.386115
city_lon = -80.741998

# Funcin para calcular distancia
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Calcular distancia
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)

# Crear mapa base
map_city = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Marcador del sitio de lanzamiento
folium.Marker([launch_site_lat, launch_site_lon], popup='Launch Site').add_to(map_city)

# Marcador de la ciudad
folium.Marker([city_lat, city_lon], popup='Closest City').add_to(map_city)

# Etiqueta de distancia
city_marker = folium.Marker(
    [city_lat, city_lon],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_city:.2f} KM</b></div>',
    )
)
map_city.add_child(city_marker)

# Lnea entre la ciudad y el sitio de lanzamiento
city_line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [city_lat, city_lon]],
    weight=2,
    color='blue'
)
map_city.add_child(city_line)

map_city
```

```python
import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Coordenadas del sitio de lanzamiento (ejemplo: CCAFS SLC-40)
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# Coordenadas del ferrocarril ms cercano (ejemplo aproximado)
rail_lat = 28.57205
rail_lon = -80.58528

# Funcin para calcular distancia
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Calcular la distancia
distance_rail = calculate_distance(launch_site_lat, launch_site_lon, rail_lat, rail_lon)

# Crear el mapa base
final_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)

# Agregar marcador del sitio de lanzamiento
folium.Marker([launch_site_lat, launch_site_lon], popup='Launch Site').add_to(final_map)

# Agregar marcador con etiqueta de distancia
rail_marker = folium.Marker(
    [rail_lat, rail_lon],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12px; color:#0077be;"><b>{distance_rail:.2f} KM</b></div>',
    )
)
final_map.add_child(rail_marker)

# Dibujar lnea azul
rail_line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [rail_lat, rail_lon]],
    weight=2,
    color='blue'
)
final_map.add_child(rail_line)

# Mostrar el mapa
final_map
```
